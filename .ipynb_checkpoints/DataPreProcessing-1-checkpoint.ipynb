```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data/sample_dataset.csv')

df.head(10)
```

```python
categorical_columns = df.select_dtypes(include = ['object', 'category', 'bool']).columns
categorical_columns
```

```python
numerical_columns = df.select_dtypes(exclude = ['object', 'category', 'bool']).columns
numerical_columns
```

```python
from sklearn.impute import SimpleImputer

cleaner = SimpleImputer(strategy = 'mean')

mean_columnData = df.iloc[:,:10]
mean_columnData
```

```python
cleaner.fit(mean_columnData)
```

```python
cleaner.transform(mean_columnData)
```

```python
df.iloc[:,:10] = cleaner.transform(mean_columnData)
df
```

```python
df = pd.read_csv('./data/sample_dataset.csv')

df.head(10)
```

```python
import warnings
warnings.filterwarnings('ignore')

df['mean area'].mean()
df['mean area'].fillna(df['mean area'].mean(), inplace=True)
df.head(10)
```

```python
X = df[['area error']]
X
```

```python
X.isnull().sum()
```

```python
cleaner = SimpleImputer(strategy = 'most_frequent')
cleaner.fit_transform(X)
```

```python
cleaner = SimpleImputer(strategy = 'constant', fill_value='undefined')
cleaner.fit_transform(X)
```

```python
from sklearn.impute import KNNImputer

cleaner = KNNImputer(n_neighbors=5)

X = df.iloc[:,:3] 
cleaner.fit_transform(X)
```

```python
from sklearn.compose import ColumnTransformer

df = pd.read_csv('./data/sample_dataset.csv')

cleaner = ColumnTransformer([
    ('numerical_transformer', SimpleImputer(strategy='mean'), numerical_columns),
    ('categorical_transformer', SimpleImputer(strategy='most_frequent'), categorical_columns)
])

cleaner.fit_transform(df)
```

# Encoding the categorical variables

```python
from sklearn.preprocessing import OneHotEncoder

cv = np.array([['A'],['A'],['B'],['D'],['E']])
cv
```

```python
enc = OneHotEncoder()
enc.fit_transform(cv).todense()
```

```python
cv = np.array([['A','X'],['B','Y'],['D','Y'],['E','Z']])
enc = OneHotEncoder(categories=[['A','B','D','E'], ['X','Y','Z']])

enc.fit_transform(cv).todense()
```

```python
from sklearn.preprocessing import OrdinalEncoder

cv = np.array([['High'],['Low'],['Medium']])

enc = OrdinalEncoder()
enc.fit_transform(cv)
```

```python
enc = OrdinalEncoder(categories = [['Low', 'Medium', 'High']])
enc.fit_transform(cv)
```

```python
from sklearn.preprocessing import LabelEncoder

cv = np.array(['A','B','C'])
enc = LabelEncoder()

enc.fit_transform(cv)
```
